```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

conv=1.634e-11 # conversion from erg to number counts of lya photons for Jly
d1=np.loadtxt('LyaUVB0216.dat')
y1=d1[1]
x1=d1[0]
d2=np.loadtxt('LyaHM12.dat')
y2=d2[1]
x2=d2[0]


# read in data
d21_hi=np.loadtxt('/Users/verag/Research/Repositories/pmfs/inputs/global_evolution_zetaIon31.50_Nsteps40_zprimestepfactor1.020_zetaX1.0e+56_alphaX1.2_TvirminX1.0e+04_Pop3_300_200Mpc__hiFSTAR')
y21_hi=d21_hi[:,6][::-1]
x21_hi=d21_hi[:,0][::-1]
d21_lo=np.loadtxt('/Users/verag/Research/Repositories/pmfs/inputs/global_evolution_zetaIon31.50_Nsteps40_zprimestepfactor1.020_zetaX1.0e+56_alphaX1.2_TvirminX1.0e+04_Pop3_300_200Mpc__loFSTAR')
y21_lo=d21_lo[:,6][::-1]
x21_lo=d21_lo[:,0][::-1]
d21_mid=np.loadtxt('/Users/verag/Research/Repositories/pmfs/inputs/global_evolution_zetaIon31.50_Nsteps40_zprimestepfactor1.020_zetaX1.0e+56_alphaX1.2_TvirminX1.0e+04_Pop3_300_200Mpc__midFSTAR'
)
y21_mid=d21_mid[:,6][::-1]
x21_mid=d21_mid[:,0][::-1]
                  
     
        
# plot models
plt.semilogy(x2,y2/conv,lw=1,label='Haard+Madau 2012')
plt.semilogy(x1,y1/conv,lw=1,label='new model (UVB)')  

plt.semilogy(x21_lo,y21_lo,'--',lw=3,label='21cmfast: LO FSTAR (lower lim)')
plt.semilogy(x21_hi,y21_hi,'--',lw=3,label='21cmfast: HI FSTAR (upper lim)')
plt.semilogy(x21_mid,y21_mid,lw=3,color='k',label='21cmfast: MID FSTAR (fiducial)')                  

plt.xlabel('z')
plt.ylabel(r'$J_{Ly\alpha} [cm^{-2}sec^{-1}Hz^{-1}sr^{-1}]$')

plt.legend(loc='lower left')
plt.savefig('Lya_models.pdf')
```
